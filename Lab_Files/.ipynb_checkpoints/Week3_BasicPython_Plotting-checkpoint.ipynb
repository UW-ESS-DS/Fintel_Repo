```python
# generate random fields
from numpy import random

N = 500 #Number of samples
#create
#Uniform distribution
x1=2*random.rand(N)-1 #Centered at 0, varying between -1 and 2
#Gaussian Distribution
x2=random.randn(N) #standard deviaton of 1
#power law
a=2 # power of distribution
x3=random.power(a,N)
#poisson
aa=4
x4 = random.poisson(aa,N)
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,5))
ax1.plot(x1);ax1.set_title('Uniform')
ax2.plot(x2);ax2.set_title('Gaussian')
ax3.plot(x3);ax3.set_title('Power')
ax4.plot(x4);ax4.set_title('Poisson')

fig2, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,5))
ax1.hist(x1,bins=N);
ax2.hist(x2,bins=N);
ax3.hist(x3,bins=N);
ax4.hist(x4,bins=N);
```

## 2D arrays in Numpy

Read CSV File

```python
# now we are going to open and read each line of a csv file
# at each line, we append the earthwuake information to their list
import csv
# first we initialize the python lists for basic earthquake properties
with open ('pnsn_catalog.csv','r') as file:
    reader = list(csv.reader(file,delimiter=','))
    L = len(reader)
    #sanity check
    print("there are " + str(L)+ " earthquakes in the catalog")
    print(f"there are {L} earthquakes in the catalog")
    mag=np.zeros(L);depth=np.zeros(L)
    lat=np.zeros(L);lon=np.zeros(L)
    for i, row in enumerate(reader[1:]):
        mag[i]=row[1]
        lat[i]=row[-4]
        lon[i]=row[-3]
        depth[i]=row[-2]
```

```python
plt.hist(mag,bins=20);plt.grid(True);plt.title("historgrams of magnitude distribution in PNW")
plt.xlabel('Local Magnitude')
```

```python
plt.scatter(lon[lon!=0],lat[lat!=0])
```

```python
import csv
with open('quake_catalog_montague_island.csv','r') as file:
    reader = list(csv.reader(file,delimiter =','))
    L = len(reader)
    print(f"there are {L} earthquakes in this catalog")
    mag = np.zeros(L);depth=np.zeros(L)
    lat = np.zeros(L);lon=np.zeros(L)
    for i, row in enumerate(reader[1:]):
        mag[i] =row[4]
        lat[i]=row[1]
        lon[i]=row[2]
        depth[i]=row[3]
```

```python
plt.hist(mag,bins=40);plt.grid(True)
fig=plt.figure()
plt.scatter(lon[lon!=0],lat[lat!=0]);plt.grid(True)
```
